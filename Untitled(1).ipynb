# Uma introdução à DifferentialEquations.jl

Seja a seguinte equação diferencial ordinária:

$$\dfrac{du}{dt} = f(u,p,t),$$

onde $p$ são os parâmetros do modelo, $t$ é a variável do tempo e $f$ é um modelo não linear de como $u$ muda com relação ao tempo. Para ter um problema de valor inicial, precisamos da condição inicial, ou seja,

$$u(t_0) = u_0.$$


## Primeiro modelo: Crescimento exponencial

Neste modelo, temos que a taxa de mudança de $u$ é proporcional ao valor corrente de $u$:

$$\dfrac{du}{dt} = au,$$

no qual temos como valor inicial $u(0)=u_0.$

Por exemplo: Digamos que colocamos 1 dólar em Bitcoin, que está aumentando a uma taxa de $98\%$ ao ano.
Tomando agora $t_0=0$ e medindo o tempo em anos, nosso modelo é:

$$\dfrac{du}{dt} = 0.98u,$$ e $u(0)=1.0.$

Neste caso, $f(u,p,t)= 0.98u,$ com $u_0=1.0.$

Tomando $t$ de forma que $t\in[0.0,1.0]$, definimos o problema como segue:

```julia
import Pkg
Pkg.add("DifferentialEquations")
Pkg.add("Plots")
Pkg.add("PyPlot")
```

```julia
using DifferentialEquations

f(u,p,t) = 0.98u
u0 = 1.0
tspan = (0.0,1.0)
prob = ODEProblem(f, u0, tspan)
println(prob)
```

Para resolver o problema, façamos:

```julia
sol = solve(prob)
```

## Analisando a solução

Afim de analisarmos a solução obtida, utilizaremos Plots.jl para mostrar o gráfico da solução, como segue:

```julia
using Plots
pyplot()
```

```julia
plot(sol)
```

Podemos nomear os eixos, legendar o gráfico e dar um título para o gráfico:

```julia
plot(sol, linewidth=3, title="Solução para a EDO linear", xaxis="Tempo (t)", yaxis="u(t)", label="Solução Obtida")
```

Usando o comando "plot!", podemos adicionar outras funções ao gráfico.
Neste caso, sabemos que a solução para a EDO é

$$u(t)=u_0e^{at}.$$

Então, adicionando essa função ao gráfico, temos:

```julia
plot!(sol.t, t->1.0*exp(0.98t), lw=3, ls=:dash, label="Solução Exata")
```

No comando anterior, "sol.t" é o vetor de pontos de tempo na qual a solução foi salva, assim como "sol.u" é o vetor de pontos da solução $u$ que foram salvos.

Temos também que "sol.u\[i\]" é a solução $u$ no tempo "sol.t\[i\]".

```julia
sol.t
```

```julia
sol.u
```

```julia
sol.t[3]
```

```julia
sol.u[3]
```

Podemos manipular os vetores da solução "sol" usando uma configuração como a que segue, que faz a soma dos vetores u e t:

```julia
[t+u for (u,t) in tuples(sol)]
```

Como os pontos são interpolados por um polinômio de Hermite de ordem 3, a solução é tratada como contínua, assim podemos obter o valor da solução em pontos intermediários para $t$ no intervalo $[0.0,1.0]$, como por exemplo $t=0.33$, com o seguinte comando:

```julia
sol(0.33)
```

## Controlando o solver

O solver possui algumas opções adicionais que podem ser passadas como parâmetro.

```julia
sol = solve(prob, abstol=1e-8, reltol=1e-8)
```

Ao diminuir o a tolerância, o número de etapas que o solucionador teve que seguir foi 9 ao invés das 5 anteriores, o que pode acarretar numa perda de velocidade para encontrar a solução.

```julia
plot(sol)
plot!(sol.t, t->1.0*exp(0.98t), lw=3, ls=:dash, label="Solução Exata")
```

Quando "saveat" é utilizado, as variáveis de saída contínua não são mais salvas, e a interpolação "sol(t)" é apenas de primeira ordem.
Passando os valores em "saveat", podemos ter uma grade irregular de pontos (pontos quaisquer dentro do intervalo).

```julia
sol = solve(prob, saveat=0.1)
```

```julia
sol = solve(prob, saveat=[0.35,0.71,0.99])
```

Por padrão, os primeiro e último valores do intervalo são salvos, mas isso é algo que pode sser alterado fazendo:

```julia
sol = solve(prob, saveat=[0.35,0.71,0.99], save_start=false, save_end=false)
```

Se precisarmos diminuir a quantidade de pontos salvos, podemos "desativar" a saída contínua, fazendo "dense = false":

```julia
sol = solve(prob, dense=false)
```

E para não salvar os pontos intermediários, basta usar "save\_everystep = false":

```julia
sol = solve(prob, save_everystep=false)
```

## Escolhendo algoritmos de solução

Quando chamamos "solve(prob)", a biblioteca "adivinha" um bom algoritmo para solucionar o problema, dadas as propriedades passadas por parâmetro no "solve", como tolerâncias, informações de salvamento, etc. 
Mas há como controlar qual algoritmo da biblioteca será utilizado.

```julia
sol = solve(prob, alg_hints=[:stiff])
```

Algoritmos rígidos precisam resolver equações implícitas e sistemas lineares em cada etapa, então, devem ser usados apenas quando necessário.

```julia
sol = solve(prob, Tsit5(), reltol=1e-6)
```

## Sistemas de EDOs: A equação de Lorentz

"Atrator de borboletas", que gerou a teoria do caos.
É definido pelo seguinte sistema de EDOs:

\begin{align}
\dfrac{dx}{dt}&= \sigma(y-x) \\
\dfrac{dy}{dt}&= x(\rho-z)-y \\
\dfrac{dz}{dt}&= xy-\beta z
\end{align}

Para definir um sistema de equações diferenciais usando a biblioteca, definimos nossa $f$ como uma função vetorial com um vetor de condições iniciais. Portanto, para o vetor $u=[x,y,z]$, temos a função derivada:

```julia
function lorentz!(du,u,p,t)
    s,r,b = p
    du[1] = s*(u[2]-u[1])
    du[2] = u[1]*(r-u[3])-u[2]
    du[3] = u[1]*u[2]-b*u[3]
end
```

Note que aqui utilizamos o formato "in-loco" que grava a saída no vetor pré-alocado $du$. Para sistemas, esse formato é mais rápido. Usaremos aqui a condição inicial $u_0=[1.0,0.0,0.0]$.

```julia
u0 = [1.0,0.0,0.0]
```

Neste modelo, usamos os parâmetros $p$. Precisamos definir esse valor no ODEProblem também. 

Usaremos aqui os parâmetros: $\sigma = 10$, $\rho = 28$ e $\beta = \frac{8}{3}$, ou seja, $s=10$, $r=28$ e $b=\frac{8}{3}$. 

Dessa forma, $p=\left[10,28,\frac{8}{3}\right]$.

Agora, geramos o ODEProblem. Adicionamos, neste caso, os parâmetros ao final. Resolveremos para $t\in[0.0,100.0]$. 

```julia
p = (10,28,8/3)
tspan = (0.0,100.0)
prob = ODEProblem(lorentz!, u0, tspan, p)
sol = solve(prob)
```

Como antes, aqui "sol.t" e "sol.u" armazenam o tempo e a solução, respectivamente.

```julia
sol.t[10],sol.u[10]
```

Além disso, a solução atua como uma matriz $sol[i,j]$, na qual $i$ é o índice referentte ao tempo $t[i]$ e $j$ é o índice referente à solução $u[i]$. 

Podemos obter uma matriz real realizando uma conversão. Apenas transformamos $sol[i,j]$ em uma matriz $A[i,j]$.

```julia
sol[3,5]
```

```julia
A = convert(Array, sol)
```

Mostrando o gráfico, teremos a série temporal de cada variável dependente:

```julia
plot(sol)
```

Se quisermos mostrar o retrato de fase, basta fazer:

```julia
plot(sol, vars=(1,2,3))
```

Por padrão os pontos são interpolados, mas podemos vê-los sem a interpolação, basta fazer:

```julia
plot(sol, vars=(1,2,3), denseplot=false)
```

Podemos ver apenas a série temporal de alguma das variáveis dependentes também, já que o tempo é o índice 0 em "vars":

```julia
plot(sol, vars=(0,3))
```

## Uma DSL para funções parametrizadas

Em muitos casos, podemos definir várias funções com parâmetros. Existe uma linguagem específica de domínio (DSL) definida pela macro "@ode\_def" para ajudar com esse problema.

Vamos observar o exemplo das equações de Lotka-Volterra, que se dá por:

\begin{align}
\dfrac{dx}{dt}&= ax - bxy \\
\dfrac{dy}{dt}&= - cy + dxy
\end{align}

```julia
function lotka_volterra!(du,u,p,t)
    du[1] = p[1]*u[1] - p[2]*u[1]*u[2]
    du[2] = -p[3]*u[2] + p[4]*u[1]*u[2]
end
```

Contudo, o código pode ficar muito carregado e difícil de acompanhar. Ao invés disso, pode-se usar a macro "@ode\_def":

```julia
lv! = @ode_def LotkaVolterra begin
    dx = a*x - b*x*y
    dy = -c*y + d*x*y
end a b c d
```

Podemos usar então o resultado como as EDOs definidas anteriormente:

```julia
u0 = [1.0,1.0]
p = (1.5,1.0,3.0,1.0)
tspan = (0.0,10.0)
prob = ODEProblem(lv!,u0,tspan,p)
sol = solve(prob)
plot(sol)
```

A sintaxe DSL não é apenas conveniente, mas também faz alguma "mágica" por trás. As vezes os solucionadores precisam usar o jacobiano nos cálculos. O DSL usa diferenciação simbólica para derivar automaticamente a função em questão.

O DSL pode derivar muitas funções. Essa habilidade é usada para acelerar os solucionadores. Uma extensão para DifferentialEquations.jl, Latexify.jl permite extrair essas partes como expressões LaTeX.

##  Tipos internos

Quandos usamos valores de ponto flutuante de 64 bits para a condição inicial, isso significa que os valores internos serão resolvidos com Float64. Se o tempo também for especificado como Float64, significa que as etapas de tempo serão Float64. Mudando esses tipos, podemos mudar o que é utilizado internamente para se obter a solução.

```julia
A = [1. 0  0 -5
     4 -2  4 -3
    -4  0  0  1
     5 -2  2  3]
u0 = rand(4,2)
tspan = (0.0,1.0)
f(u,p,t) = A*u
prob = ODEProblem(f,u0,tspan)
sol = solve(prob)
```

Não há diferença real entre o que foi feito anteriormente, mas agora $u$ é uma matriz $4\times 2.$

```julia
sol[3]
```

Nesta biblioteca, podemos usar qualquer tipo que defina as quatro operações e tenha uma norma apropriada.

Por exemplo, se quisermos números de ponto flutuante de precisão arbitrária, podemos alterar a entrada para ser uma matriz de "BigFloat":

```julia
big_u0 = big.(u0)
```

E podemos solucionar o ODEProblem com números de precisão arbitrária, usando essa condição inicial:

```julia
prob = ODEProblem(f,big_u0,tspan)
sol = solve(prob)
```

```julia
sol[2,3]
```

Para realmente fazer uso disso, gostaríamos de mudar a precisão em "abstol" e "reltol" para serem pequenos. 

Observe que o tipo "tempo" é diferente com relação aos tipos das variáveis dependentes, e isso pode ser usado para otimizar o algoritmo mantendo várias precisões.

Também podemos converter o tempo em precisão arbitrária:

```julia
prob = ODEProblem(f,big_u0,big.(tspan))
sol = solve(prob)
```

Para vetores pequenos, geralmente é mais rápido fazer  operações em vetores estáticos por meio do pacote StaticArrays.jl. A sintaxe é semelhante à dos vetores normais, mas para esses vetores usamos a macro "@SMatrix" para indicar que queremos criar um vetor estático.

```julia
Pkg.pkg"add StaticArrays"
using StaticArrays
A = @SMatrix [ 1.0  0.0 0.0 -5.0
               4.0 -2.0 4.0 -3.0
              -4.0  0.0 0.0  1.0
               5.0 -2.0 2.0  3.0]
u0 = @SMatrix rand(4,2)
tspan = (0.0,1.0)
f(u,p,t) = A*u
prob = ODEProblem(f,u0,tspan)
sol = solve(prob)
```

```julia
sol[3]
```
