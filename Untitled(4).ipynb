# DiffEqFlux.jl - Uma biblioteca Julia para Equações Diferenciais Neurais

## O que as equações diferenciais têm a ver com aprendizado de máquina?

Existem três maneiras comuns de defini uma transformação não linear: modelagem direta, aprendizado de máquina e equações diferenciais.

Escrever diretamente a função não linear só funciona se conhecemos a forma funcional exata que relaciona a entrada e a saída do problema. Entretato, em muitos casos, essas relações exatas não são previamente conhecidas.

Uma maneira de resolver o problema é usando aprendizado de máquina. Em um problema típico de aprendizado de máquina, recebemos uma entrada $x$ e desejamos prever uma saída $y$. O modelo que gera a saída $y$ é um modelo de aprendizado de máquina, digamos $ML$. Daí, durante o treinamento tentamos ajustar os parâmetros de $ML$ para que ele gere previsões suficientemente precisas. Podemos usar $ML$ para inferência (produzir $y$s para novas entradas $x$). Nesse caso, $ML$ é apenas uma transformação não linear $y=ML(x)$. A razão que torna $ML$ interessante é que seu formato é básico, mas ela se adapta aos dados fornecidos.

Por exemplo, uma rede neural simples (em forma de matriz) com funções de ativação sigmóide é  multiplicações de matrizes seguidas pela aplicação de funções sigmóides. De forma que

$$ML(x)=\sigma(W_3\cdot\sigma(W_2\cdot\sigma(W_1\cdot x)))$$

é uma rede neural profunda de três camadas, na qual $W=(W_1,W_2,W_3)$ são parâmetros a serem aprendidos. Daí, escolhemos $W$ de forma que $ML(x)=y$ se ajuste de forma razoável à função desejada.
A teoria e a prática de aprendizado de máquina afirmam que esta é uma boa maneira de aprender não linearidades.

Por exemplo, o Teorema de Aproximação Universal afirma que, para camadas ou parâmetros suficientes (para matrizes $W_i$ suficientemente grandes), $ML(x)$ pode aproximar qualquer função linear suficientemente próxima. Mas há ressalvas, sendo uma delas que se tem que aprender tudo sobre a transformação não linear diretamente dos dados. Em muitos casos, não conhecemos a equação não linear completa, mas podemos saber detalhes sobre sua estrutura.

Por exemplo, para um modelo de crescimento populacional de coelhos, digamos que queremos aprender

$$coelhos\,\,amanh\tilde{a}=Modelo(coelhos\,\,hoje)$$

Neste caso, temos um conhecimento à priori de que a taxa de nascimento depende da população atual. A maneira de estabelecer matematicamente essa suposição estrutural é por meio de uma equação diferencial. Aqui, estamos dizendo que a taxa de natalidade da população de coelhos em um determinado momento aumenta quando temos mais coelhos. A maneira mais simples de codificar matematicamente isso é

$$coelhos'(t)=\alpha\cdot coelhos(t)$$

onde $\alpha$ é alguma constante aprendida. 

Equações diferenciais são essencialmente equações de "como as coisas mudam" e, portanto, "onde as coisas estarão" é a solução para uma equação diferencial.

Assim, à medida que nossos modelos de aprendizado de máquina crescem e necessitam de maiores quantidades de dados, as equações diferenciais se tornam uma opção atraente para especificar não linearidades de uma forma que possa ser aprendida, mas restrita. Elas são essencialmente uma forma de incorporar conhecimento prévio específico do domínio das relações estruturais entre entrada e saída. Dada essa visão sobre ambos os métodos, ambos possuem vantagens e desvantagens, tornando-os ferramentas complementares para modelagem.

## O que é a Equação Diferecial Neural Ordinária?

É uma das muitas maneiras de pôr ambos os assuntos (EDOs e redes neurais) juntos. De maneira simples, ao invés de aprender a transformação não linear diretamente, desejamos aprender as estruturas da transformação não linear. Ao invés de fazer $y=ML(x)$, colocamos o modelo de aprendizagem de máquina na derivada, $y'=ML(x)$, e então resolvemos a EDO.

Uma motivação para isso é que definir o modelo dessa forma, e em seguida resolver a EDO usando o método mais simples sujeito a erros, o método de Euler, o que obtemos é equivalente a uma rede neural residual.
A forma com a qual o método de Euler funciona é baseada no fato de $y'(x)=\frac{dy}{dx}$, portanto,

$$\Delta y=(y_{n+1}-y_{n})=\Delta x\cdot ML(x),$$
e daí,
$$y_{n+1}=y_n+\Delta x\cdot ML(x_n).$$

Sua estrutura é semelhante à uma rede neural residual (modelos profundos de ResNet efetivamente se aproximam de um tipo de modelo "infinitamente profundo", já que cada camada tende à zero).
Ao invés de adicionar mais camadas, podemos apenas modelar a equação diferencial diretamente e, em seguida, resolvê-la usando um solucionador de EDO específico.

## Como se resolve uma EDO?

A ideia é definir um ODEProblem por meio de uma equação derivada $u'=f(u,p,t)$, fornecer uma condição inicial $u_0$, um intervalo de tempo $tspan$ para resolver e especificar os parâmetros $p$.

Por exemplo, para as equações de Lotka-Volterra, que descrevem a dinâmica de coelhos e lobos, e podem ser escritas como:

\begin{align}
x'&=\alpha x+\beta xy\\
y'&=-\delta y+\gamma xy
\end{align}

escrevemos em Julia como:

```julia
import Pkg
Pkg.add("DifferentialEquations")
Pkg.add("Flux")
Pkg.add("DiffEqFlux")
Pkg.add("Plots")
```

```julia
using DifferentialEquations
function lotka_volterra(du,u,p,t)
    x, y = u
    a,b,d,g = p
    du[1] = dx = a*x - b*x*y
    du[2] = dy = - d*y + g*x*y
end
u0 = [1.0,1.0]
tspan = (0.0,10.0)
p = [1.5,1.0,3.0,1.0]
prob = ODEProblem(lotka_volterra, u0, tspan, p)
```

Então, para resolver a equação diferencial, podemos simplesmente chamar "solve" para resolver "prob":

```julia
sol = solve(prob)

using Plots
plot(sol)
```

Além disso, podemos fazer com que nossa condição inicial ($u_0$) e intervalos de tempo ($tspan$) sejam funções dos parâmetros (os elementos de $p$). Por exemplo, podemos definir o ODEProblem:

```julia
u0_f(p,t0) = [p[2],p[4]]
tspan_f(p) = (0.0,10*p[4])
p = [1.5,1.0,3.0,1.0]
prob = ODEProblem(lotka_volterra, u0_f, tspan_f, p)
```

Ou seja, $$u_0\_f(p,t_0)=(\beta,\gamma)$$
e $$tspan\_f(p)=(0.0,10\gamma)$$
Desta forma, tudo sobre o problema é determinado pelo vetor de parâmetros ($p=(\alpha,\beta,\delta,\gamma)$).

## Vamos colocar uma EDO em uma estrutura de rede neural!

Para entender a incorporação de uma EDO em uma rede neural, veremos o que é realmente uma camada de rede neural. Uma camada é, na verdade, apenas uma função diferenciável que leva um vetor de tamanho $n$ e produz um novo vetor de tamanho $m$.

Acontece que solucionadores de equações diferenciais também se encaixam nessa estrutura: uma solução leva em algum vetor $p$ (que pode incluir parâmetros como o ponto de partida inicial) e produz algum novo vetor, a solução. Além disso, é diferenciável, o que significa que podemos colocá-lo em um "programa" diferenciável maior. Este "programa" maior pode incluir redes neurais, e podemos continuar usando técnicas de ootimização padrão como ADAM para otimizar seus pesos.

DiffEqFlux.jl torna conveniente fazer exatamente isso. Por exemplo, resolvendo uma equação como antes, sem gradientes, temos:

```julia
p = [1.5,1.0,3.0,1.0]
prob = ODEProblem(lotka_volterra, u0, tspan, p)
sol = solve(prob, Tsit5(), saveat=0.1)
A = sol[1,:]
```

Obtemos o seguinte resultado:

```julia
plot(sol)
t = 0:0.1:10.0
scatter!(t,A)
```

O solve já cuida do tipo de tratamento necessário para tornar a solução compatível com o framework da rede neural (aqui Flux). Para mostrar isso, definiremos uma rede neural com a função como nossa única camada, e em seguida, uma função de perda que é a distância ao quadrado dos valores de saída da função constante 1. Em Flux, isso se parece com:

```julia
using Flux
p = [2.2,1.0,2.0,0.4]
params = Flux.params(p)

function predict_rd()
    solve(prob, Tsit5(),p=p,saveat=0.1)[1,:]
end

loss_rd() = sum(abs2,x-1 for x in predict_rd())
```

Agora dizemos à Flux para treinar a rede neural executando por um período 100 para minimizar nossa função perda "loss\_rd()" e, assim, obter os parâmetros otimizados:

```julia
data = Iterators.repeated((),100)
opt = ADAM(0.1)
cb = function()
    display(loss_rd())
    display(plot(solve(remake(prob,p=p), Tsit5(), saveat=0.1), ylim=(0,6)))
end
cb()
Flux.train!(loss_rd, params, data, opt, cb=cb)
```

A Flux encontra os parâmetros da rede neural ($p$) que minimizam a função perda, ou seja, treina a rede neural: acontece que a passagem para a frente da rede neural inclui a resolução de uma EDO. Uma vez que nossa função de perda punha uma penalidade sempre que o número de coelhos estava longe de 1, nossa rede neural encontrou parâmetros onde nossa população de coelhos e lobos são constantes igual à 1.

Agora temos a solução de EDOs como apenas uma camada, podemos adicioná-la em qualquer lugar. Por exemplo, o perceptron multicamadas é escrito em Flux como:

```julia
m = Chain(
  Dense(28^2, 32, relu),
  Dense(32, 10),
  softmax)
```

E se tivéssemos uma EDO apropriada, que tivesse um vetor de parâmetros do tamanho certo, podemos colocá-la ali:

```julia
m = Chain(
  Dense(28^2, 32, relu),
  p->solve(prob, Tsit5(), p=p, saveat=0.1)[1,:],
  Dense(32,10),
  softmax)
```

Ou podemos inseri-la em uma rede neural convolucional, onde as camadas anteriores definem a condição inicial para a EDO:

```julia
m = Chain(
  Conv((2,2), 1=>16, relu),
  x -> maxpool(x,(2,2)),
  Conv((2,2),16=>8, relu),
  x -> maxpool(x, (2,2)),
  x -> reshape(x, :, size(x, 4)),
  x -> solve(prob, Tsit5(), u0=x, saveat=0.1)[1,:],
  Dense(288,10),softmax)
```

## Por que um pacote de solucionador de EDO completo é necessário para que isso funcione bem?

Considere o seguinte exemplo, a EDO Rober. A implementação mais bem testada (e otimizada) de um método Adams-Bashforth-Moulton é o integrador CVODE no pacote C ++ SUNDIALS. Vamos usar DifferentialEquations.jl para chamar CVODE com seu método de Adams e fazer com que ele resolva o ODE para nós: 

```julia
rober = @ode_def Rober begin
    dy1 = - k1*y1 + k3*y2*y3
    dy2 = k1*y1 - k2*y2^2 - k3*y2*y3
    dy3 = k2*y2^2
end k1 k2 k3

prob = ODEProblem(rober, [1.0;0.0;0.0], (0.0,1e11), (0.04,3e7,1e4))
solve(prob, CVODE_Adams())
```

Este e o método "dopri" (Fortran) perdem o controle e não conseguem resolver a equação. Isso acontece porque a EDO é rígida, e portanto, métodos com regiões de estabilidade menores não serão capazes de resolvê-la adequadamente. Por outro lado, o método "KenCarp4()" para este problema, resolve a equação:

```julia
sol = solve(prob, KenCarp4())
plot(sol,xscale=:log10, tspan=(0.1,1e11))
```

## Que tipos de equações diferenciais existem?

As EDOs são apenas um tipo de equação diferencial. Existem muitos recursos adicionais que podemos adicionar à estrutura de uma equação diferencial. Por exemplo, a quantidade de coelhos no futuro não depende do número de coelhos agoras porque leva uma quantidade de tempo diferente de zero para um pai chegar ao termo depois que um filho é iniciado. Assim, a taxa de natalidade dos coelhos é, na verdade, devido à quantidade de coelhos no passado. Usar um termo de atrasona derivada de uma equação diferencial torna essa equação conhecida como equação diferencial de atraso (EDA). Como DifferentialEquations.jl lida com EDAs por meio da mesma interface que EDOs, ela também pode ser usada como uma camada no Flux. 

```julia
function delay_lotka_volterra(du,u,h,p,t)
    x, y = u
    a, b, d, g = p
    du[1] = dx = (a - b*y)*h(p,t-0.1)[1]
    du[2] = dy = (d*x - g)*y
end
h(p,t) = ones(eltype(p),2)
prob = DDEProblem(delay_lotka_volterra, [1.0,1.0],h,(0.0,10.0), constant_lags=[0.1])
p = [2.2,1.0,2.0,0.4]
params = Flux.params(p)

Pkg.add("DiffEqSensitivity")
using DiffEqSensitivity
function predict_rd_dde()
    solve(prob, MethodOfSteps(Tsit5()),p=p,sensealg=TrackerAdjoint(),saveat=0.1)[1,:]
end
loss_rd_dde() = sum(abs2,x-1 for x in predict_rd_dde())
loss_rd_dde()
```

Além disso, podemos adicionar aleatoriedade à nossa equação diferencial para simular como eventos aleatórios podem causar nascimentos extras ou mais mortes do que o esperado. Esse tipo de equação é conhecido como equação diferencial estocástica (EDE). Como DifferentialEquations.jl lida com EDEs, eles podem ser tratados como uma camada no Flux de maneira semelhante. Aqui está uma camada de rede neural com uma EDE:

```julia
function lotka_volterra_noise(de,u,p,t)
    du[1] = 0.1u[1]
    du[2] = 0.1u[2]
end
prob = SDEProblem(lotka_volterra,lotka_volterra_noise,[1.0,1.0],(0.0,5.0))
p = [2.2,1.0,2.0,0.4]
params = Flux.params(p)
function predict_sde()
    solve(prob,SOSRI(),p=p,sensealg=TrackerAdjoint(),saveat=0.1,abstol=1e-1,reltol=1e-1)[1,:]
end
loss_rd_sde() = sum(abs2,x-1 for x in predict_sde())
loss_rd_sde()
```

E podemos treinar a rede neural para vê-la em ação e encontrar parâmetros para tornar a quantidade de coelhos próxima de constante:

```julia
data = Iterators.repeated((), 100)
opt = ADAM(0.1)
cb = function()
    display(loss_rd_sde())
    display(plot(solve(remake(prob,p=p),SOSRI(),saveat=0.1),ylim=(0,6)))
end

cb()
Flux.train!(loss_rd_sde, params, data, opt, cb=cb)
```

Existem equações diferenciais que são constantes por partes usadas em simulações biológicas, ou de difusão de salto de modelos financeiros, e os solucionadores mapeiam diretamente para a estrutura da rede neural de Flux por meio de DiffEqFlux.jl.

## Implementando a camada de rede neural EDO em Julia

Vamos implementar a camada neural EDO em Julia. Lembre-se de que esta é simplesmente uma EDO em que a função derivada é definida pela própria rede neural. Para fazer isso, vamos primeiro definir a rede neural para a derivada. Na Flux, podemos definir um perceptron multicamadas com 1 camada oculta e uma função de ativação tangente hiperbólica como:

```julia
dudt = Chain(Dense(2,50,tanh), Dense(50,2))
```

Para definir uma camada neural EDO, precisamos apenas dar a ela um intervalo de tempo e usar a função NeuralODE:

```julia
tspan = (0.0f0,25.0f0)
node = NeuralODE(dudt,tspan,Tsit5(),saveat=0.1)
```

Para executar isso na GPU, é suficiente fazer com que a condição inicial e a rede neural estejam na GPU. Assim, todas as operações internas ocorrerão na GPU.

```julia
node = NeuralODE(gpu(dudt),tspan,Tsit5(),saveat=0.1)
```

## Entendendo o comportamento da camada neural EDO por exemplo

Vamos usar a camada de EDO neural em um exemplo para descobrir o que ela significa. Primeiro, vamos gerar uma série temporal de uma EDO em pontos de tempo uniformemente espaçados. Usaremos a equação de teste do artigo Neural ODE.

```julia
u0 = Float32[2.;0.]
datasize = 30
tspan = (0.0f0,1.5f0)

function trueODEfunc(du,u,p,t)
    true_A = [-0.1 2.0; -2.0 -0.1]
    du .= ((u.^3)'true_A)'
end
t = range(tspan[1],tspan[2],length=datasize)
prob = ODEProblem(trueODEfunc,u0,tspan)
ode_data = Array(solve(prob,Tsit5(),saveat=t))
```

Agora vamos comparar uma EDO neural com esses dados. Para fazer isso, definiremos uma rede neural de camada única que tem apenas a mesma EDO neural de antes.

```julia
dudt = Chain(x -> x.^3, Dense (2,50,tanh), Dense(50,2))
n_ode = NeuralODE(dudt,tspan,Tsit5(),saveat=t,reltol=1r-7,abstol=1e-9)
ps = Flux.params(n_ode)
```
